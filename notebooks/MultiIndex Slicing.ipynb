```python
pd.set_option('max_rows',14)
from IPython.display import display as disp
```

```python
# create the test data
def mklbl(prefix,n):
    return ["%s%s" % (prefix,i)  for i in range(n)]

miindex = MultiIndex.from_product([mklbl('A',4),
                                   mklbl('B',2),
                                   mklbl('C',4),
                                   mklbl('D',2)],
                                  names=['first','second','third','fourth']) 

micolumns = MultiIndex.from_tuples([('a','foo'),('a','bar'),
                                    ('b','foo'),('b','bah')],
                                     names=['lvl0', 'lvl1'])


dfmi = DataFrame(np.arange(len(miindex)*len(micolumns)).reshape((len(miindex),len(micolumns))),
                 index=miindex,
                 columns=micolumns).sortlevel().sortlevel(axis=1)

dfmi
```

```python
# sorting
dfmi.index.lexsort_depth
```

```python
dfmi.sortlevel(level='third')
```

```python
dfmi.sortlevel(level='third').index.lexsort_depth
```

```python
dfmi.xs('C1',level='third',drop_level=False)
```

```python
dfmi.loc[(slice(None),slice(None),'C1'),:]
```

```python
# new syntax
idx = pd.IndexSlice

dfmi.loc[idx[:,:,'C1',:],:]
```

```python
dfmi.loc[idx[:,:,['C1','C2'],:],:]
```

```python
dfmi.loc[idx[:,:,['C1','C2'],'D0'],:]
```

```python
dfmi.loc[idx[:,:,['C1','C2'],'D0'],'a']
```

```python
dfmi.loc[idx[:,:,['C1','C2'],'D0'],idx[:,'foo']]
```
